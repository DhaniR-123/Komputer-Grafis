```python
import numpy as np
import matplotlib.pyplot as plt

plt.grid(True)
plt.axis('on')

plt.axis([-10, 10, -10, 10])  # Format axis: [xmin, xmax, ymin, ymax]
plt.gca().set_aspect('equal')  # Menggunakan gca() untuk mengatur aspek rasio

# --------- custom grid ---------
x1 = -10
x2 = 10
y1 = 10
y2 = -10
dx = 0.5
dy = -0.5

for x in np.arange(x1, x2, dx):
    for y in np.arange(y2, y1, -dy):  # perhatikan urutan y agar sesuai arah sumbu
        plt.scatter(x, y, s=1, color='lightgray')

# --------- square box ---------
plt.plot([-5, 5], [-5, -5], linewidth=2, color='k')   # bawah
plt.plot([5, 5], [-5, 5], linewidth=2, color='k')     # kanan
plt.plot([5, -5], [5, 5], linewidth=2, color='k')     # atas
plt.plot([-5, -5], [5, -5], linewidth=2, color='k')   # kiri

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.axis([0,150,100,0])
r=40
alpha1=np.radians(0)
alpha2=np.radians(360)
dalpha=np.radians(2)
xc=75
yc=50
plt.scatter(xc,yc,s=10,color='k')
for alpha in np.arange(alpha1,alpha2,dalpha):
    x=xc+r*np.cos(alpha)
    y=yc+r*np.sin(alpha)
    plt.scatter(x,y,s=5,color='k')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.axis([-10,140,90,-10])

plt.axis('on')
plt.grid(True)

plt.title('Sample Axes')

# ---------------------- grid
dx=5
dy=-5
for x in np.arange(x1,x2,dx):
    for y in np.arange(y1,y2,dy):
        plt.scatter(x,y,s=1,color='lightgray')

# ---------------------- coordinate axes
plt.arrow(0,0,20,0,head_length=4,head_width=3,color='k')
plt.arrow(0,0,0,20,head_length=4,head_width=3,color='k')

plt.show()
```
